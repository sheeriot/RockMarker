```python
import folium
import pandas as pd

# Read the input file containing marker data
data = pd.read_csv('sites/eaglelake/rocks.csv')

# Set the center coordinates
center_lat, center_lon = 44.683,-76.7

# Create a Folium map object with the specified center
map = folium.Map(location=[center_lat, center_lon],zoom_start=14)

# Iterate over the rows of the data and add markers to the map
for index, row in data.iterrows():
    lat = row['latitude']
    lon = row['longitude']
    name = row['nickname']
    marker = folium.Marker(location=[lat, lon], tooltip=name)
    marker.add_to(map)

# Display the map
map
```
